```python
#creating simple text file
import os
os.makedirs("../data/text_files", exist_ok=True)
```

```python
with open("../data/text_files/intro.txt", "w", encoding="utf-8") as f:
    f.write("""
Python is a high-level, interpreted programming language known for its
simplicity and readability. It uses an easy-to-understand syntax that
allows developers to write fewer lines of code compared to many other
programming languages.

One of the key features of Python is its dynamic typing. This means
you do not need to declare variable types explicitly, making development
faster and more flexible.

Python supports multiple programming paradigms, including procedural,
object-oriented, and functional programming. This makes it suitable for
a wide variety of applications.

Another important feature is Python’s vast standard library. It provides
built-in modules for tasks such as file handling, mathematical operations,
web development, and data processing.

Python is platform-independent, meaning the same code can run on
Windows, macOS, and Linux without modification.

Python also has a large and active community. This results in a rich
ecosystem of third-party libraries and frameworks used in fields like
web development, data science, artificial intelligence, and automation.

Because of its simplicity and powerful features, Python is widely used
by beginners as well as experienced developers.

""")
```

```python
### Textloader
from langchain_community.document_loaders import TextLoader

loader =  TextLoader("../data/text_files/intro.txt" , encoding="utf-8")
document = loader.load()
print(document)
```

```python
from langchain_community.document_loaders import PyMuPDFLoader , PyPDFLoader , DirectoryLoader

loader = DirectoryLoader(
    "../data/pdf",
    glob="*.pdf",
    loader_cls=PyMuPDFLoader
)

documents = loader.load()

print(documents)
```
